**Mutable Types**

Mutable types are those whose values can be changed after they are created. This means you can modify the internal state of the object without creating a new object. Examples of mutable types in Python include:

* Lists: `list`
* Dictionaries: `dict`
* Sets: `set`


**Immutable Types**

Immutable types, on the other hand, are those whose values cannot be changed after they are created. When you modify an immutable object, a new object is created with the updated value. Examples of immutable types in Python include:

* Integers: `int`
* Floats: `float`
* Strings: `str`
* Tuples: `tuple`

**Arithmetic expressions**

Arithmetic expressions involve mathematical operations using operators such as addition (+), subtraction (-), multiplication (*), division (/), floor division (//), modulus (%), and exponentiation (**).

Evaluation Process

Python follows the order of operations (PEMDAS/BODMAS) when evaluating arithmetic expressions:

* Parentheses/Brackets: Expressions within parentheses are evaluated first.
* Exponents: Exponentiation operations are performed next.
* Multiplication and Division: These operations have equal precedence and are evaluated from left to right.

* Addition and Subtraction: These operations also have equal precedence and are evaluated from left to right.

Example 1: Integer Arithmetic

````python
x = 10
y = 5

# Addition
sum_result = x + y  # sum_result will be 15

# Subtraction
diff_result = x - y  # diff_result will be 5

# Multiplication
prod_result = x * y  # prod_result will be 50

# Division (results in a float)
div_result = x / y  # div_result will be 2.0

# Floor Division (results in an integer)
floor_result = x // y  # floor_result will be 2

# Modulus (remainder)
mod_result = x % y  # mod_result will be 0

# Exponentiation
exp_result = x ** y  # exp_result will be 100000

print('sum_result = ', sum_result)
print('diff_result = ', diff_result)
print('prod_result = ', prod_result)
print('div_result = ', div_result)
print('floor_result = ', floor_result)
print('mod_result = ', mod_result)
print('exp_result = ', exp_result)
````

Example 2: Mixed Integer and Float Arithmetic

````python
a = 7
b = 3.5

# Addition (results in a float)
sum_result = a + b  # sum_result will be 10.5

# Subtraction (results in a float)
diff_result = a - b  # diff_result will be 3.5

# Multiplication (results in a float)
prod_result = a * b  # prod_result will be 24.5

# Division (results in a float)
div_result = a / b  # div_result will be 2.0

#floor divison (results in a float)
floor_result = a // b  # floor_result will be 2.0

print('sum_result = ', sum_result)
print('diff_result = ', diff_result)
print('prod_result = ', prod_result)
print('div_result = ', div_result)
print('floor_result = ', floor_result)
````

**Relational expressions**

Relational expressions, also known as comparison expressions, use comparison operators to compare two values and determine the relationship between them. The result of a relational expression is always a boolean value (True or False).


**Comparison Operators**

Here are the comparison operators available in Python:


```
Operator	Description
==	      Equal to
!=	      Not equal to
>	       Greater than
<       	Less than
>=      	Greater than or equal to

```


Evaluation Process

When evaluating a relational expression, Python compares the values on either side of the operator and returns True if the relationship holds true, otherwise False.

Examples

````python
x = 10
y = 5

# Equal to
result1 = x == y

# Not equal to
result2 = x != y

# Greater than
result3 = x > y

# Less than
result4 = x < y

# Greater than or equal to
result5 = x >= y

# Less than or equal to
result6 = x <= y
````



**Logical expressions**

Logical expressions combine multiple conditions using logical operators to produce a single boolean result (True or False). They are commonly used in conditional statements and loops to control the flow of execution based on the truthiness of multiple conditions.

Logical Operators

Python provides three logical operators:

```
Operator	Description
and	     Returns True if both operands are True, otherwise False
or	      Returns True if at least one operand is True, otherwise False
not     	Returns the opposite boolean value of the operand
```
Evaluation Process

Logical expressions are evaluated based on the precedence and associativity of the logical operators. The not operator has the highest precedence, followed by and, and then or. All logical operators are left-associative, meaning they are evaluated from left to right.

**Type casting**
Type casting, also known as type conversion, is the process of changing the data type of a value from one type to another. It allows you to perform operations that require specific data types or to ensure compatibility between different parts of your code.

Type Casting Functions

Python provides several built-in functions for type casting:


```
int(): Converts a value to an integer.
float(): Converts a value to a floating-point number.
str(): Converts a value to a string.
bool(): Converts a value to a boolean.
```



````python
# Convert a string to an integer
num_str = "123"
num_int = int(num_str)  # num_int will be 123

# Convert an integer to a float
num_int = 45
num_float = float(num_int)  # num_float will be 45.0

# Convert a float to a string
num_float = 3.14
num_str = str(num_float)  # num_str will be "3.14"

# Convert a number to a boolean
num = 0
num_bool = bool(num)  # num_bool will be False (0 is considered False)
````

**Statement flow control**



Statement flow control refers to the order in which statements are executed in a program.

Compound statement :
A group of statement executed as a unit is called a compound statemnet. The conditional and loop statements are compound statements.

Python provides several control flow statements that allow you to alter the execution sequence:

1. Conditional Statements: These statements execute a block of code only if a certain condition is true. The main conditional statements are:
* if statement: Executes a block of code if a condition is true.
* if-else statement: Executes one block of code if a condition is true, and another block if it's false.
* if-elif-else statement: Allows you to check multiple conditions and execute different blocks of code accordingly.

````python
# Example:
age = 20

if age >= 18:
  print("You are eligible to vote.")
else:
  print("You are not eligible to vote yet.")
````

2. Looping Statements: These statements execute a block of code repeatedly as long as a condition is true. The main looping statements are:
* for loop: Iterates over a sequence (like a list, tuple, or string) and executes a block of code for each item in the sequence.
* while loop: Executes a block of code as long as a condition is true.

````python
# Example:
numbers = [1, 2, 3, 4, 5]

for number in numbers:
  print(number * 2)

i = 0
while i < 5:
  print(i)
  i += 1
````

For Loop

The for loop is used to iterate over a sequence (such as a list, tuple, string, or range) or other iterable object. Here's the basic syntax:

````python
for item in sequence:
    # Code to execute for each item in the sequence
````

Reasoning:

The loop iterates through each item in the sequence. For each item, the code within the loop block is executed.

````python
# Iterate over a list of numbers
numbers = [1, 2, 3, 4, 5]

for number in numbers:
    print(number)

# Iterate over a string
message = "Hello"

for letter in message:
    print(letter)

# Iterate using the range function
for i in range(1, 6):  # Numbers from 1 to 5
    print(i)
````

While Loop
The while loop is used to repeatedly execute a block of code as long as a condition is True. Here's the basic syntax:

````python
while condition:
    # Code to execute while the condition is True
````

Reasoning:

Before each iteration, the condition is evaluated. If it's True, the code within the loop block is executed. If it's False, the loop terminates.

````python
# Count from 1 to 5
count = 1

while count <= 5:
    print(count)
    count += 1

# Loop until a condition is met
number = 0

while number < 10:
    number = int(input("Enter a number greater than or equal to 10: "))

print("You entered:", number)
````

Choosing Between For and While

* Use a for loop when you know the number of iterations in advance or need to iterate over a specific sequence.
* Use a while loop when the number of iterations is uncertain and depends on a condition.


Infinite Loops:

Be careful with while loops, as they can lead to infinite loops if the condition is never False. Make sure there's a way for the condition to become False within the loop.

Jump statements

Jump statements are used to alter the flow of execution in a program by transferring control to a different location. In Python, the jump statements are break, continue, and pass

Break Statement

The break statement is used to exit a loop prematurely, regardless of the loop's condition or the number of iterations remaining. When break is encountered within a loop, the loop terminates immediately, and the program execution continues with the code following the loop.

Reasoning:

break is often used when you want to stop the loop based on a specific condition being met, even if the loop hasn't finished iterating through all its elements.

Example

````python
for i in range(10):
  if i == 5:
    break
  print(i)
````

In this example, the loop iterates through numbers from 0 to 9. However, when i becomes 5, the break statement is executed, causing the loop to terminate immediately.

Continue Statement

The continue statement is used to skip the rest of the current iteration in a loop and proceed to the next iteration. When continue is encountered within a loop, the remaining code within the loop block for the current iteration is skipped, and the loop proceeds to the next iteration.

Reasoning:

continue is often used when you want to selectively ignore certain elements or conditions within the loop without terminating the loop entirely.

Example

````python
for i in range(10):
  if i % 2 == 0:
    continue
  print(i)
````

In this example, the loop iterates through numbers from 0 to 9. However, when i is even (i % 2 == 0), the continue statement is executed, skipping the print(i) statement for that iteration and proceeding to the next iteration.

pass

Acts as a placeholder statement.
It doesn't perform any action but is used to maintain the structure of the code, such as in an empty loop or function.

````python
for i in range(5):
    pass  # Do nothing in this loop
    print(i)
````

Loop Else Statements

In Python, loops can have an optional else block. The code inside the else block is executed only if the loop completes normally without encountering a break statement.

Reasoning:

Loop else statements can be useful when you want to perform an action after the loop has finished iterating through all its elements, but only if the loop wasn't prematurely terminated by a break.

Example

````python
for i in range(5):
  if i == 3:
    break
  print(i)
else:
  print("Loop completed normally")
````

In this example, the loop iterates through numbers from 0 to 4. However, when i becomes 3, the break statement is executed, and the loop terminates without executing the else block.

If the break statement was removed or the condition was never met, the loop would complete all iterations, and the else block would be executed, printing "Loop completed normally"

Nested Loops

You can nest loops inside other loops to create more complex iteration patterns. In a nested loop, the inner loop completes all its iterations for each iteration of the outer loop.

Reasoning:

Nested loops are commonly used when you need to iterate over multiple sequences or perform operations on multidimensional data structures like matrices or tables.

Example

````python
for i in range(3):
  for j in range(2):
    print(f"i: {i}, j: {j}")
````

In this example, the outer loop iterates three times (for i values 0, 1, and 2), and for each iteration of the outer loop, the inner loop iterates twice (for j values 0 and 1). The print statement inside the inner loop is executed for every combination of i and j values.
